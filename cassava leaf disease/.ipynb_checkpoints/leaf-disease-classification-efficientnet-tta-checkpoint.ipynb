```python
# 환경과 관련된 부분을 고정시키는 함수
def seed_everything(seed):
    random.seed(seed) # seed 고정
    os.environ['PYTHONHASHSEED'] = str(seed) 
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = True
    
# img load
def get_img(path):
    im_bgr = cv2.imread(path) # 지정된 경로에서 이미지 read
    im_rgb = im_bgr[:, :, ::-1] # 여기서 배열안의 요소를 거꾸로 불러온 이유는? rgb값
    #print(im_rgb)
    return im_rgb
```

# Dataset
- 우리가 모델에 넣을 데이터셋을 나타내는 추상화된 클래스
- __init__: 초기값 세팅, transform을 할당할 수 있음
- __len__: 데이터셋의 크기 리턴
- __getitem__: i번째 샘플(데이터 인덱스 위치)를 찾는데 사용, 이미지 판독
    - 데이터 형태는 주로 이미지명, class로 구축

```python
class CassavaDataset(Dataset):
    def __init__(
        self, df, data_root, transforms=None, output_label=True
    ):
        
        super().__init__()
        self.df = df.reset_index(drop=True).copy()
        self.transforms = transforms # transformer 불러오기
        self.data_root = data_root # 이미지 파일 경로
        self.output_label = output_label # output 라벨링
    
    def __len__(self):
        return self.df.shape[0]
    
    def __getitem__(self, index: int):
        
        # get labels
        if self.output_label: # output label != None
            target = self.df.iloc[index]['label']
          
        # 이미지 파일 경로 (data_root: 기본경로, 이미지 파일 이름)
        path = "{}/{}".format(self.data_root, self.df.iloc[index]['image_id'])
        
        # 위에서 만든 이미지값을 불러옴
        img  = get_img(path)
        
        # transformer로 이미지 전처리
        if self.transforms:
            img = self.transforms(image=img)['image']
            
        # do label smoothing
        if self.output_label == True:
            return img, target
        else:
            return img
```

# Train, validation image augmentations 재정의
- parameter example: http://incredible.ai/pytorch/2020/04/25/Pytorch-Image-Augmentation/

```python
# train transformer
def get_train_transforms():
    return Compose([
            RandomResizedCrop(CFG['img_size'], CFG['img_size']),# 이미지 크롭 사이즈 변환(여기선 512x512)
            Transpose(p=0.5),
            # resize: padding?
            HorizontalFlip(p=0.5), # 좌우 반전(0.5는 디폴트 1을 넣어야 반전)
            VerticalFlip(p=0.5), # 상하 반전 (0.5는 디폴트 그대로)
            # Affine(x): 회전, 평행이동, 스케일, 반전을 모두 포함한 변환
            # Grayscale(p=): 이미지 흑백처리
            # Perspective(): color, gray random
            # Rotation: 이미지 각도 변환
            # randomchoice: 여러 transforms 파라미터를 삽입하여 랜덤성을 반영할 수 있음
            ShiftScaleRotate(p=0.5),
            HueSaturationValue(hue_shift_limit=0.2, sat_shift_limit=0.2, val_shift_limit=0.2, p=0.5),
            RandomBrightnessContrast(
                brightness_limit=(-0.1,0.1), contrast_limit=(-0.1, 0.1), p=0.5
            ), # 밝기변환(colorjitter: jitter 내에도 많은 파라미터가 존재함)
            Normalize(
                mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225],
                max_pixel_value=255.0,
                p=1.0
            ),# 글을 봤는데 이미지 정규화처리는 totensor 이후에 사용하는 것이 좋지만 v2에서는 다른가? 
            # 보통 정형화된 데이터 셋에서 정규화는 데이터의 분포때문에 사용하는 것인데 이미지에서 사용하는 이유는?
            CoarseDropout(p=0.5),
            Cutout(p=0.5),
            ToTensorV2(p=1.0),# 채널의 위치를 바꿔줌 (h,w,c) -> (c,h,w)
        ], p=1.)
  
# validation transformer      
def get_valid_transforms():
    return Compose([
            CenterCrop(CFG['img_size'], CFG['img_size'], p=1.),
            Resize(CFG['img_size'], CFG['img_size']),
            Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0),
            ToTensorV2(p=1.0),
        ], p=1.)

# test transformer
def get_inference_transforms():
    return Compose([
            RandomResizedCrop(CFG['img_size'], CFG['img_size']),
            Transpose(p=0.5),
            HorizontalFlip(p=0.5),
            VerticalFlip(p=0.5),
            HueSaturationValue(hue_shift_limit=0.2, sat_shift_limit=0.2, val_shift_limit=0.2, p=0.5),
            RandomBrightnessContrast(brightness_limit=(-0.1,0.1), contrast_limit=(-0.1, 0.1), p=0.5),
            Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0),
            ToTensorV2(p=1.0),
        ], p=1.)
```

# Model

- efficientnet pretrained model

```python
class CassvaImgClassifier(nn.Module):
    def __init__(self, model_arch, n_class, pretrained=False):
        super().__init__()
        self.model = timm.create_model(model_arch, pretrained=pretrained)# 모델 생성
        n_features = self.model.classifier.in_features# 모델이 분류할 클래스 수 정의
        # torch로 생성한 neuralnet
        self.model.classifier = nn.Linear(n_features, n_class)# 이미지 차원 정의
        
    def forward(self, x):
        x = self.model(x)
        return x
```

# Main Loop

```python
def inference_one_epoch(model, data_loader, device):
    model.eval()

    image_preds_all = []
    
    pbar = tqdm(enumerate(data_loader), total=len(data_loader))
    for step, (imgs) in pbar:
        imgs = imgs.to(device).float()
        
        image_preds = model(imgs)   #output = model(input)
        image_preds_all += [torch.softmax(image_preds, 1).detach().cpu().numpy()]
        
    
    image_preds_all = np.concatenate(image_preds_all, axis=0)
    return image_preds_all
```

```python
if __name__ == '__main__':
     # for training only, need nightly build pytorch

    seed_everything(CFG['seed'])
    
    # 층화 교차검증셋 분할: 5분할
    folds = StratifiedKFold(n_splits=CFG['fold_num']).split(np.arange(train.shape[0]), train.label.values)
    
    for fold, (trn_idx, val_idx) in enumerate(folds):
        # we'll train fold 0 first
        if fold > 0:
            break 

        print('Inference fold {} started'.format(fold))
        
        # 1. dataset으로 img load, img transform 해서 불러오기
        valid_ = train.loc[val_idx,:].reset_index(drop=True)
        valid_ds = CassavaDataset(valid_, '../input/cassava-leaf-disease-classification/train_images/', transforms=get_inference_transforms(), output_label=False)
        
        test = pd.DataFrame()
        test['image_id'] = list(os.listdir('../input/cassava-leaf-disease-classification/test_images/'))
        test_ds = CassavaDataset(test, '../input/cassava-leaf-disease-classification/test_images/', transforms=get_inference_transforms(), output_label=False)
        
        # Dataset를 통해서 만들어진 것을 반복해주는 iterator의 역할
        val_loader = torch.utils.data.DataLoader(
            valid_ds, 
            batch_size=CFG['valid_bs'],
            num_workers=CFG['num_workers'],
            shuffle=False,
            pin_memory=False,
        )
        
        tst_loader = torch.utils.data.DataLoader(
            test_ds, 
            batch_size=CFG['valid_bs'],
            num_workers=CFG['num_workers'],
            shuffle=False,
            pin_memory=False,
        )
        
        # gpu할당
        device = torch.device(CFG['device'])
        # model 생성
        model = CassvaImgClassifier(CFG['model_arch'], train.label.nunique()).to(device)
        
        val_preds = []
        tst_preds = []
        
        #for epoch in range(CFG['epochs']-3):
        for i, epoch in enumerate(CFG['used_epochs']):   
            # 학습된 가중치 불러오는 부분
            model.load_state_dict(torch.load('../input/pytorch-efficientnet-baseline-train-amp-aug/{}_fold_{}_{}'.format(CFG['model_arch'], fold, epoch)))
            
            with torch.no_grad():
                for _ in range(CFG['tta']):
                    val_preds += [CFG['weights'][i]/sum(CFG['weights'])/CFG['tta']*inference_one_epoch(model, val_loader, device)]
                    tst_preds += [CFG['weights'][i]/sum(CFG['weights'])/CFG['tta']*inference_one_epoch(model, tst_loader, device)]

        val_preds = np.mean(val_preds, axis=0) 
        tst_preds = np.mean(tst_preds, axis=0) 
        
        print('fold {} validation loss = {:.5f}'.format(fold, log_loss(valid_.label.values, val_preds)))
        print('fold {} validation accuracy = {:.5f}'.format(fold, (valid_.label.values==np.argmax(val_preds, axis=1)).mean()))
        
        del model
        torch.cuda.empty_cache()
```

```python
test['label'] = np.argmax(tst_preds, axis=1)
test.head()
```
