## ***3. 방문통계***

```python
'''
year : 1부터 2020년
month : 1부터 1월 (1 ~ 12)
'''

def ymd_visit_statistic_crawler(year,month):
    base_url = 'https://www.jamietshop.co.kr/admin/statistic_visitor/visitor_basic'
    browser = webdriver.Chrome('/Users/sinmin-yong/GitHub/studies/Riot_DB구축/chromedriver')

    browser.get(base_url)
    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[1]/tbody/tr[2]/td/div/input')[0].send_keys('jamiet') # id
    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[1]/tbody/tr[4]/td/div/input')[0].send_keys('woaxmghkdlxld12!@') # pw
    time.sleep(0.5)
    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[2]/tbody/tr[6]/td/input')[0].click() # login click

    time.sleep(2)

    #시간별 클릭
    browser.find_elements_by_xpath('//*[@class="slc-body"]//form//label[3]')[0].click()

    time.sleep(0.5)

    #연도 선택
    browser.find_elements_by_xpath('//*[@class="slc-body"]//form//select')[0].\
            find_elements_by_xpath('option')[year].click()

    #월 선택
    browser.find_elements_by_xpath('//*[@class="slc-body"]//form//select')[1].\
            find_elements_by_xpath('option')[month].click()

    #일 선택(무조건 첫 검색 전에는 31일까지 존재)
    day_click_ls = browser.find_elements_by_xpath('//*[@class="slc-body"]//form//select')[2].\
            find_elements_by_xpath('option')[1:]
    
    if year ==1:
        y = '2020'
    elif year == 2:
        y = '2019'
    elif year == 3:
        y = '2018'

    visit_time_summary = pd.DataFrame()

    day_click_ls[0].click()

    #검색 클릭
    browser.find_elements_by_xpath('//*[@class="slc-body"]//form//div/span[@class="btn small"]')[0].click()
    time.sleep(1)
    
    day_limit = len(browser.find_elements_by_xpath('//*[@class="slc-body"]//form//select')[2].\
                            find_elements_by_xpath('option'))
        
    for day in range(2,day_limit+1):

        
        html = browser.page_source
        soup = BeautifulSoup(html,'html.parser')

        #원하는 날짜의 테이블
        table = soup.find('div',{'id':'visitor_hourly_table'})

        #컬럼 수집
        '''
        col_ls = table.find_all('th')
        column_name = []


        for i in col_ls:
            column_name.append(i.get_text(strip = True))
        '''

        #내용수집
        value_ls = table.find_all('tr')[1:]


        visit_time_summary00 = pd.DataFrame()

        for j in value_ls:
            td_ls = j.find_all('td')

            hour = td_ls[0].get_text(strip = True)
            pv = float(re.sub('[^.0-9]','',td_ls[1].get_text(strip=True)))
            visit_count = float(re.sub('[^.0-9]','',td_ls[2].get_text(strip=True)))
            pv_per_visit = float(re.sub('[^.0-9]','',td_ls[3].get_text(strip=True)))
            ymd = y + '-' + str(month) + '-' +str(day-1)

            visit_time_summary0 = pd.DataFrame({'hour' : [hour],
                                                'page_view' : [pv],
                                                'visit_count' : [visit_count],
                                                'pv_per_visit' : [pv_per_visit],
                                                'date' : [ymd]})

            visit_time_summary00 = visit_time_summary00.append(visit_time_summary0)

        visit_time_summary = visit_time_summary.append(visit_time_summary00)
        
        if day == day_limit:
            break
            
        else:
            browser.find_elements_by_xpath('//*[@class="slc-body"]//form//select')[2].\
                find_elements_by_xpath('option')[day].click()
            
            
            #검색 클릭
            browser.find_elements_by_xpath('//*[@class="slc-body"]//form//div/span[@class="btn small"]')[0].click()
            time.sleep(1)


    browser.close()
    browser.quit()
    
    return visit_time_summary
```

## ***6. 주문리스트***
    - 10일씩 수집

```python
def order_list_crawler(start_date,end_date):
    
    base_url = 'https://www.jamietshop.co.kr/admin/order/catalog?query_string=&no=&keyword_type=&keyword=&regist_date_type=&date_field=regist_date&regist_date%5B%5D={}&regist_date%5B%5D={}&chk_step%5B75%5D=1&referer='
    
    browser = webdriver.Chrome('/Users/mindragon/Nrise python/chromedriver')

    browser.get(base_url.format(start_date,end_date))

    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[1]/tbody/tr[2]/td/div/input')[0].send_keys('jamiet') # id
    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[1]/tbody/tr[4]/td/div/input')[0].send_keys('woaxmghkdlxld12!@') # pw
    time.sleep(0.5)
    browser.find_elements_by_xpath('//*[@id="loginForm"]/table/tbody/tr/td/div[2]/table[2]/tbody/tr[6]/td/input')[0].click() # login click
    
    time.sleep(2)
    body = browser.find_element_by_tag_name('body')#스크롤하기 위해 소스 추출

    num_of_pagedowns = 0
    #10번 밑으로 내리는 것
    while True:

        if num_of_pagedowns != 0 and num_of_pagedowns % 10 == 0:
            html = browser.page_source
            soup = BeautifulSoup(html,'html.parser')

            table = soup.find('tbody',{'class':'ltb order-ajax-list'})

            row_ls = table.find_all('tr',{'class':re.compile('list-row step75 important')})

            idx = int(row_ls[-1].find_all('td')[2].get_text(strip = True))

            if idx == 1:
                print('Full Scroll')
                break
            else:
                body.send_keys(Keys.PAGE_DOWN)
                time.sleep(0.5)
                num_of_pagedowns += 1


        else:
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.5)
            num_of_pagedowns += 1
            
    html = browser.page_source
    soup = BeautifulSoup(html,'html.parser')

    table = soup.find('tbody',{'class':'ltb order-ajax-list'})
    
    row_ls = table.find_all('tr',{'class':re.compile('list-row step75 important')})
    

    browser.close()
    browser.quit()
    
    order_df = pd.DataFrame()
    
    for i in range(len(row_ls)):
        order_ls = row_ls[i].find_all('td')

        # 주문날짜
        order_date = order_ls[4].get_text(strip = True)
        #주문번호
        order_number = order_ls[6].get_text()
        #받는자/주문자
        user_name = order_ls[10].get_text(strip = True)
        #결제날짜
        purchase_date = row_ls[0].find_all('td')[11].get_text(strip = True)
        #결제금액
        price = int(re.sub('[^0-9]','',order_ls[12].get_text(strip = True)))

        order_df0 = pd.DataFrame({'order_date': [order_date],
                                  'order_number' : [order_number],
                                  'user_name' : [user_name],
                                  'purchase_date' : [purchase_date],
                                  'price' : [price]})

        order_df = order_df.append(order_df0)
    
        
    return order_df
```
