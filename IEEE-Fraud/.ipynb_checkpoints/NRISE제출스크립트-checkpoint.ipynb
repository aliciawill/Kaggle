```python
'''사용함수'''
def age_interval(x):
    if x >=40:
        return '40대'
    elif 30<=x<40:
        return '30대'
    elif 20<=x<30:
        return '20대'
    else:
        return '10대'
    
def point_interval(x):
    if x>=90:
        return '90점대'
    elif 80<=x<90:
        return '80점대'
    elif 70<=x<80:
        return '70점대'
    elif 60<=x<70:
        return '60점대'
    elif 50<=x<60:
        return '50점대'
    elif 40<=x<50:
        return '40점대'
    elif 30<=x<40:
        return '30점대'
    elif 20<=x<30:
        return '20점대'
    elif 10<=x<20:
        return '10점대'
    else:
        return '10점미만'
    
def height_interval(x):
    if x>=190:
        return '190-'
    elif 180<=x<190:
        return '180-190'
    elif 170<=x<180:
        return '170-180'
    elif 160<=x<170:
        return '160-170'
    elif 150<=x<160:
        return '150-160'
    else:
        return '-140'
def dis_interval(x):
    if 150000<=x:
        return '150-'
    elif 50000<=x<150000:
        return '50-150'
    elif 10000<=x<50000:
        return '10-50'
    else:
        return '-10'
```
